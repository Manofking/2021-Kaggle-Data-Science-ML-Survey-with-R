```r
kaggle_survey_2021_responsess$gender_new <- ifelse(kaggle_survey_2021_responsess$Q2=="Man", "Male", #sas의 if cv_death=“Cardiac” then CV_DEATH=1; else if cv_death=“Vascular” then CV_DEATH=1;개념.
                                                   ifelse(kaggle_survey_2021_responsess$Q2=="Woman", "Female",
                                                          ifelse(kaggle_survey_2021_responsess$Q2=="Nonbinary", "Others","Others")))
```

```r
table(kaggle_survey_2021_responsess$gender_new) 
names(kaggle_survey_2021_responsess)
```

```r
##=========================1.3.1 성별 분포 확인.

kaggle_survey_2021_responsess %>% 
  dplyr::group_by(gender_new) %>% 
  summarize(N = n()) %>% 
  ggplot(aes(gender_new, N)) + 
  geom_col() +
  geom_text(aes(label = N), size = 5, vjust = 1.2, color = "#FFFFFF") +  
  ggtitle("Distribution of sex") +
  labs(x = "Gender", y = "Count")
```

```r
target <- data.frame(table(kaggle_survey_2021_responsess$gender_new))
target$por <- target$Freq/sum(target$Freq)*100


# 막대 그래프 생성
ggplot(target, aes(x ="", y = por, fill = Var1)) +
  geom_col()

ggplot(target, aes(x ="", y = por, fill = Var1)) + 
  geom_col() +
  coord_polar(theta = "y") +
  geom_text(aes(label = paste0(Var1, "\n", round(por, 2), "%")), 
            position = position_stack(vjust = 0.5))
```

```r
##=========================1.3.2 연령별 분포 확인.
kaggle_survey_2021_responsess %>% 
  dplyr::group_by(Q1) %>%  
  summarize(N = n()) %>% 
  ggplot(aes(Q1, N)) + 
  geom_col() +
  geom_text(aes(label = N), size = 5, vjust = 1.2, color = "#FFFFFF") +  
  ggtitle("Distribution of Age") +
  labs(x = "Age", y = "Count")
```

```r
kaggle_survey_2021_responsess %>% 
  dplyr::group_by(Q1, gender_new) %>% 
  summarize(N = n()) %>%
  ggplot(aes(x = Q1, y = N, fill = gender_new )) +
  geom_col(position = "dodge") + 
  geom_text(aes(label = N), position = position_dodge(width=0.5))
```

```r
kaggle_survey_2021_responsess %>% 
  dplyr::group_by(Q1, gender_new) %>%  
  summarize(N = n()) %>% 
  ggplot(aes(Q1, N, fill = gender_new)) + geom_col() +  
  #geom_text(aes(label = N), size = 5, vjust = 1.2, color = "#FFFFFF") + ggtitle("Bar plot of Sex") + 
  labs(x = "Sex", y = "Count")

```

```r
kaggle_survey_2021_responsess %>% 
  group_by(Q3) %>% 
  summarise(n = n()) %>% 
  arrange(desc(n)) %>%
  head(10) %>%
  ggplot(aes(x = Q3 %>% fct_reorder(n, .desc = TRUE), 
             y = n)) +
  geom_col() +
  geom_text(aes(label = n), size = 5, vjust = 1.2, color = "#FFFFFF") + 
  ggtitle("Distribution of Country") +
  labs(x = "Country", y = "Count")
```
